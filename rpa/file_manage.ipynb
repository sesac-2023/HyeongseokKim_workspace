```python
DIR = './실습1/pdf_files/'
os.listdir(DIR)
```

```python
DIR = './실습1/pdf_files/'
pdf_list = os.listdir(DIR)[1:]
for pdf in pdf_list:
    folder_name = pdf.split('_')[1].replace('-','_')
    if os.path.isdir(DIR + folder_name):
        pass
    else:
        os.makedirs(DIR + folder_name)
    original_dir = DIR + pdf
    # ./pdf_files/TEST_2019-1-2_23-57-18.pdf
    target_dir = DIR + folder_name + '/' + pdf
    # ./pdf_files/2019_1_2/TEST_2019-1-2_23-57-18.pdf
    shutil.move(original_dir, target_dir)
```
