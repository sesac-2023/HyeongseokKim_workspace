```python
from sklearn.datasets import load_wine
```

```python
import pandas as pd
data_wine=load_wine()
```

```python
wine_df=pd.DataFrame(data_wine.data ,columns=data_wine.feature_names)
```

```python
wine_df
```

```python
import numpy as np
def euclidean_distance(a,b):
  dist=np.sqrt(np.sum((a-b)**2))
  return dist
```

```python
dist_lst=[]
d_result=[]
for i in range(len(wine_df)):
  dist_lst=[]
  for j in range(len(wine_df)):
    base_point = wine_df.iloc[i].values
    target_point = wine_df.iloc[j].values
    w_dist=euclidean_distance(base_point, target_point)
    dist_lst.append(w_dist)
  d_result.append(dist_lst)
```

```python
pd.DataFrame(d_result)
```

```python
def cos_simularity(a,b):
  cos=np.sum(a*b) / (np.sqrt(np.sum(a**2))*np.sqrt(np.sum(b**2)))
  return cos
```

```python
cos_lst=[]
c_result=[]
for i in range(len(wine_df)):
  cos_lst=[]
  for j in range(len(wine_df)):
    base_point = wine_df.iloc[i].values
    target_point = wine_df.iloc[j].values
    cos_sim=cos_simularity(base_point, target_point)
    cos_lst.append(cos_sim)
  c_result.append(cos_lst)
```

```python
pd.DataFrame(c_result)
```
