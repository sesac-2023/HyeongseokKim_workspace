```python
url='https://finance.naver.com/sise/sise_quant.nhn'
headers = {'data-useragent': 'Mozilla/5.0 (Windows NT 10.0;Win64;x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/114.0.0.0 Whale/3.21.192.18 Safari/537.36'} 
response = requests.get(url, headers=headers)
soup =bs(response.text, "html.parser")

lst=[[soup.find_all('th')[1].text, soup.find_all('th')[2].text]]
for i in range(0,100):
    name=soup.find_all('a', {'class':'tltle'})[i].text
    price=int(soup.find_all('td', {'class':'number'})[10*i].text.replace(',',''))
    lst.append([name, price])

lst
```

```python
import pandas as pd
```

```python
pd.set_option('display.max_rows', None)
stock_df=pd.DataFrame(lst[1:], columns=lst[0])
stock_df
```

```python
import time
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By 
```

```python
driver=webdriver.Chrome()
driver.get('https://www.jungle.co.kr/')
body = driver.find_element(By.TAG_NAME,'body')

title_cats=driver.find_elements(By.CLASS_NAME, 'thumb_list')
for tc in title_cats:
    count=6
    for i in range(count):
        title = tc.find_elements(By.CSS_SELECTOR, 'span.title')[i].text
        cat = tc.find_elements(By.CSS_SELECTOR,'a.thumb_cate')[i].text
        print("Title:", title)
        print("Category:", cat)
        print()
        time.sleep(2)
        continue
    more = driver.find_element(By.CLASS_NAME, 'more')
    more.click()
    continue
```
